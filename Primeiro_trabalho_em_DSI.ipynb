# Introdução à Programação - Recursão

### Iteração

```python
def soma_iterativa(n):
    soma = 0
    for i in range(1, n + 1):
        soma += i
    return soma

# Pede ao usuário para inserir o valor de n
n = int(input("Insira o valor de n: "))

# Chama a função e passa o valor inserido como argumento
resultado = soma_iterativa(n)

# Exibe o resultado
print("A soma iterativa de 1 até", n, "é:", resultado)
```

### Recursão

```python
def soma_recursiva(n):
    if n == 1:
        return 1
    else:
        return n + soma_recursiva(n - 1)

# Pede ao usuário para inserir o valor de n
n = int(input("Insira o valor de n: "))

# Chama a função e passa o valor inserido como argumento
resultado = soma_recursiva(n)

# Exibe o resultado
print("A soma recursiva de 1 até", n, "é:", resultado)
```

### Iteração

```python
def fatorial_iterativo(n):
    resultado = 1
    for i in range(2, n + 1):
        resultado *= i
    return resultado

# Pede ao usuário para inserir o valor de n
n = int(input("Insira o valor de n: "))

# Chama a função e passa o valor inserido como argumento
resultado = fatorial_iterativo(n)

# Exibe o resultado
print("O produto fatorial de", n, "é:", resultado)
```

### Iteração

```python
def fibonacci_iterativo(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

# Pede ao usuário para inserir o valor de n
a = int(input("Insira o valor de a: "))

# Chama a função e passa o valor inserido como argumento
resultado = fibonacci_iterativo(a)

# Exibe o resultado
print("A sequência de Fibonacci de", a, "é:", resultado)
```

### Recursão

```python
def fibonacci_recursivo(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursivo(n-1) + fibonacci_recursivo(n-2)

    # Pede ao usuário para inserir o valor de n
n = int(input("Insira o valor de n: "))

# Chama a função e passa o valor inserido como argumento
resultado = fibonacci_iterativo(n)

# Exibe o resultado
print("A sequência de Fibonacci de", n, "é:", resultado)
```

### Recursão Otimizada

```python
def fibonacci_recursivo_memoizado(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci_recursivo_memoizado(n-1, memo) + fibonacci_recursivo_memoizado(n-2, memo)
    return memo[n]

# Pede ao usuário para inserir o valor de n
n = int(input("Insira o valor de n: "))

# Chama a função e passa o valor inserido como argumento
resultado = fibonacci_recursivo_memoizado(n)

# Exibe o resultado
print("A sequência de Fibonacci de", n, "é:", resultado)
```

### Torre de Hanói

```python
def hanoi(n, origem, destino, auxiliar):
    if n == 1:
        print(f"Mova disco 1 de {origem} para {destino}")
        return
    hanoi(n-1, origem, auxiliar, destino)
    print(f"Mova disco {n} de {origem} para {destino}")
    hanoi(n-1, auxiliar, destino, origem)

# Pede ao usuário para inserir o número de discos
n = int(input("Insira o número de discos: "))

# Chama a função e passa o número de discos como argumento
hanoi(n, 'Início', 'Destino', 'Auxiliar')

# A função irá imprimir a sequência de movimentos
```

## Exemplo em Python de Economia e Administração

```python
def valor_futuro(valor_atual, taxa_crescimento, periodos):
    if periodos == 0:
        return valor_atual
    else:
        return valor_futuro(valor_atual * (1 + taxa_crescimento), taxa_crescimento, periodos - 1)

valor_atual = float(input("Insira o valor atual: "))
taxa_crescimento = float(input("Insira a taxa de crescimento (como decimal): "))
periodos = int(input("Insira o número de períodos: "))

resultado = valor_futuro(valor_atual, taxa_crescimento, periodos)
print(f"O valor futuro é: R$ {round(resultado,2)}")
```
